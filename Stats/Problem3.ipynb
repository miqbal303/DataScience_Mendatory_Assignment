```python
import pandas as pd
import numpy as np
import scipy.stats as stats


df = pd.read_csv('https://drive.google.com/file/d/1mCjtYHiX--mMUjicuaP2gH3k-SnFxt8Y/view?usp=share_')

# Data
blood_pressure_before = [130, 142, 120, 135, 148, 122, 137, 130, 142, 128, 135, 140, 132, 145, 124, 128, 136, 143, 127, 139, 135, 131, 127, 130, 142, 128, 136, 140, 132, 145, 124, 128, 136, 143, 127, 139, 135, 131, 127, 130, 142, 128, 136, 140, 132, 145, 124, 128, 136, 143, 127, 139, 135, 131, 127, 130, 142, 128, 136, 140, 132, 145, 124, 128, 136, 143, 127, 139, 135, 131, 127, 130, 142, 128, 136, 140, 132, 145, 124, 128, 136, 143, 127, 139, 135, 131, 127, 130, 142, 128, 136, 140, 132, 145, 124, 128, 136, 143, 127, 139, 135, 131, 127]
blood_pressure_after = [120, 135, 118, 127, 140, 118, 129, 124, 137, 125, 129, 132, 125, 136, 118, 122, 130, 139, 123, 132, 131, 126, 120, 123, 139, 122, 129, 136, 127, 140, 119, 121, 129, 137, 122, 135, 130, 125, 121, 124, 124, 139, 123, 131, 136, 127, 141, 118, 121, 129, 137, 122, 135, 130, 125, 121, 124, 124, 139, 123, 131, 136, 127, 141, 118, 121, 129, 137, 122, 135, 130, 125, 121, 124, 124, 139, 123, 131, 136, 127, 141, 118, 121, 129, 137, 122, 135, 130, 125, 121, 124, 124, 139, 123, 131, 136, 127]

df.head
```

```python
# a. Dispersion
range_before = np.max(blood_pressure_before) - np.min(blood_pressure_before)
range_after = np.max(blood_pressure_after) - np.min(blood_pressure_after)

variance_before = np.var(blood_pressure_before)
variance_after = np.var(blood_pressure_after)

std_dev_before = np.std(blood_pressure_before)
std_dev_after = np.std(blood_pressure_after)

print("Dispersion:")
print("Range (Blood Pressure Before):", range_before)
print("Range (Blood Pressure After):", range_after)
print("Variance (Blood Pressure Before):", variance_before)
print("Variance (Blood Pressure After):", variance_after)
print("Standard Deviation (Blood Pressure Before):", std_dev_before)
print("Standard Deviation (Blood Pressure After):", std_dev_after)

# b. Mean and Confidence Interval
mean_before = np.mean(blood_pressure_before)
mean_after = np.mean(blood_pressure_after)

n = len(blood_pressure_before)
std_err_before = stats.sem(blood_pressure_before)
std_err_after = stats.sem(blood_pressure_after)

confidence_interval_before = stats.t.interval(0.95, n - 1, loc=mean_before, scale=std_err_before)
confidence_interval_after = stats.t.interval(0.95, n - 1, loc=mean_after, scale=std_err_after)

print("\nMean and Confidence Interval:")
print("Mean (Blood Pressure Before):", mean_before)
print("Mean (Blood Pressure After):", mean_after)
print("Confidence Interval (Blood Pressure Before):", confidence_interval_before)
print("Confidence Interval (Blood Pressure After):", confidence_interval_after)

# c. Mean Absolute Deviation and Standard Deviation
mad_before = np.mean(np.abs(blood_pressure_before - mean_before))
mad_after = np.mean(np.abs(blood_pressure_after - mean_after))

print("\nMean Absolute Deviation:")
print("MAD (Blood Pressure Before):", mad_before)
print("MAD (Blood Pressure After):", mad_after)

print("\nStandard Deviation:")
print("Standard Deviation (Blood Pressure Before):", std_dev_before)
print("Standard Deviation (Blood Pressure After):", std_dev_after)

# d. Correlation Coefficient and Significance Test
correlation_coefficient, p_value = stats.pearsonr(blood_pressure_before, blood_pressure_after)

print("\nCorrelation Coefficient and Significance:")
print("Correlation Coefficient:", correlation_coefficient)
print("p-value:", p_value)
```
